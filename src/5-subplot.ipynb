### Exercise 5: Subplot Layout

```python
try:
    import matplotlib.pyplot as plt
    import numpy as np
except ImportError as e:
    print(f"Error: {e}. Please ensure that matplotlib and numpy are installed.")

fig, axs = plt.subplots(2, 2)

# Line chart in the first subplot
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set(xlabel='x', ylabel='sin(x)')

# Bar chart in the second subplot
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set(xlabel='Categories', ylabel='Values')

# Scatter plot in the third subplot
x = np.random.rand(10)
y = np.random.rand(10)
axs[1, 0].scatter(x, y)
for i, txt in enumerate(range(len(x))):
    axs[1, 0].annotate(txt, (x[i], y[i]))
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set(xlabel='x', ylabel='y')

# Pie chart in the fourth subplot
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')

fig.tight_layout()
plt.show()
```
